## Motivation of Stationarity in time series.

We want a time series model $X_{t}$ over time to compeletely specify a joint distribution of $(X_{1}, X_{2}, X_{3}, .... X_{t})$

**But the problem with this approach it has too many parameters** 

**Most information in first and second moment of X**

-`special case` : Multivariatee gaussian


Stationary time series

`Stictly stationary time series`

All statistical properties stay the same when shift from t to t+h

`Weakly stationay time series`

1. $E[X_{t+h}] = E[X_{t}] = X[X_{1}]$ constant not depending on t
2. $Cov(X_{t+h}, X_{t})$ only depends on **h** not **t**

## AR process

AR specifies output is dependent on previous terms and on it's own stochastic term

AR is not always stationary,

![](https://upload.wikimedia.org/wikipedia/commons/thumb/c/ce/ArTimeSeries.svg/1920px-ArTimeSeries.svg.png)

```python
from statsmodels.tsa.arima_process import arma_generate_sample
import matplotlib.pyplot as plt
```

Sample AR processes.

AR(0)

```python
_ =plt.plot(arma_generate_sample(ar=[1,0],ma=[1], nsample=100))
```

AR(1)

```python
_ =plt.plot(arma_generate_sample(ar=[1,0.9],ma=[1], nsample=100))
```
