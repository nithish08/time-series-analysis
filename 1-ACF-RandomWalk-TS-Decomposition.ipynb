## Auto Covariance function (ACVF) and Autor correlation Function (ACF)

Auto-covariance function of $X_{t}$ at lag h is given by $\gamma(h) = Covariance(X_{t+h}, X_{t})$

Autocovariance function of $X_{t}$ at lag h is given by $\rho(h) = correlation(X_{t+h}, X_{t})$

$\rho_{X}(h) =  \gamma_{X}(h) / \gamma_{X}(0)$


```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(series)
plt.show()
```

## Basic models of time series

The basic models for time series are 

1. White Noise
2. Random Walk
3. Binary Process

### White Noise

1. White Noise $X_{t} = WN(0, \sigma^{2})$


 $$ \gamma_{X}(h)=   \left\{
\begin{array}{ll}
      \sigma^{2} & h=0 \\
      0 & h\neq0 
\\
\end{array} 
\right.  $$

### Random Walk

Random Walk $S_{t} = X_{1} + X_{2} + X_{3} + ..... + X_{t}$

X is a binary process generally.


\begin{align}
\operatorname{cov}(Y_{t+h},Y_t) &= \operatorname{cov}(Y_t+W,Y_t)\\
&= \operatorname{cov}(Y_t,Y_t) + \operatorname{cov}(W,Y_t)
&{\scriptstyle\text{this step follows because the covariance operator is bilinear;}}\\
&= \operatorname{var}(Y_t) + 0 &\scriptstyle{\text{0 because
independent RVs}~W~\text{and}~Y_t~\text{have zero covariance;}}\\
&= \operatorname{var}(Y_t)\\
&= t\sigma^2
\end{align}

## Properties of Time Series

`Trend`: Consistent directional movement

`Seasonality`: regular and prectable fluctuations.

$ X_{t} = m_{t} + s_{t} + \epsilon_{t}$ -> **Additive** Model for timeseries

$ X_{t} = m_{t} * s_{t} * \epsilon_{t}$ -> **Multiplicative** Model for timeseries

observed ts = trend-component + seasonal component + error term

`Purpose` of classical decomposition of time serise is extract and estimate `Trend` and `Seasonal` component so that $\epsilon_{t}$ will be a stationary proess.

### Estimating Trend using OLSE (Polynomial fitting)

$ m_{t} = a_{0} + a_{1}t + a_{2} t^{2} + a_{3} t^{3} + .... $

choose the p with minimum residual for this model.

### Estimating Seasonal component

$s_{t} = a_{0} + a \cos(\frac{n}{d} . t) + b \sin(\frac{n}{d} . t) $

## Classical Decomposition of Time Series.

1. _Trend_ -> Moving Average Filter to estimate Trend.
2. _Seasonality_ -> Average of detrended values for that season

## Code for classical decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
#additive model decomposition
_ = plt.figure(figsize=(10,10))
result = seasonal_decompose(series, model='additive', period=1)
result.plot()
plt.show()
```
